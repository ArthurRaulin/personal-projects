```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = ['MSFT', 'AAPL', 'GOOG']
df = yf.download(tickers, start='2023-01-01', end='2023-12-31')
```

```python
df = df['Close'].dropna()

df_returns = df.pct_change().dropna()

df_cumreturns = (1+df_returns).cumprod() - 1
df_cumreturns = df_cumreturns.dropna()
```

```python
plt.plot(df_cumreturns.index, df_cumreturns['AAPL'],color='blue', label='Apple')
plt.plot(df_cumreturns.index, df_cumreturns['MSFT'], color='red', label='Microsoft')
plt.plot(df_cumreturns.index, df_cumreturns['GOOG'], color='green', label='Google')
plt.xlabel('Dates')
plt.ylabel('Rendements cumulés')
plt.title('Graphiques de rendements cumulés de trois actions américaines')
plt.legend()
plt.show()
```
